```python
mat_a_inv = np.linalg.inv(mat_a)
mat_b_faulty = np.matmul(mat_a_inv, mat_c)
mat_b_faulty
```

```python
mat_b_inv = np.linalg.inv(mat_b)
mat_a_faulty = np.matmul(mat_c, mat_b_inv)
mat_a_faulty
```

```python
def matmul_breakdown(mat_a, mat_b, index):
    j = index[0]
    k = index[1]
    rows_a = len(mat_a)
    cols_a = len(mat_a[0])
    rows_b = len(mat_b)
    cols_b = len(mat_b[0])
    if cols_a != rows_b:
        raise ValueError("Matrix A's columns must be the same as Matrix B's rows")
    result = np.zeros((rows_a, cols_b))
    for i in range(cols_a):
        result[j][k] += mat_a[j][i] * mat_b[i][k]
        print(f"result[{j}, {k}] += {mat_a[j][i]} * {mat_b[i][k]} -> {result[j, k]}")
    return result[j][k]
```

```python
matmul_breakdown(mat_a, mat_b_faulty, (0,0))
```

```python
matmul_breakdown(mat_a_faulty, mat_b, (0,0))
```
